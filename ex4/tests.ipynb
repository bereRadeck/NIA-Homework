```python
# for debugging
%matplotlib inline
from Initializer import Initializer
from DonorGenerator import DonorGenerator
from TrialGenerator import TrialGenerator
from Selector import Selector
from DE import DE
from PPP import PPP
import matplotlib.pyplot as plt
"""
Step 1: Read values of the control parameters of DE: scale
factor F, crossover rate Cr, and the population size NP from
user.
"""

# Parameter:
np=50
#xmin = 0
#xmax = 100
problemnumber = 1
F = 0.3
Cr = 0.5
iterations = 2000


#initializer = Initializer(xmin, xmax, np)
ppp = PPP(problemnumber)
initializer = Initializer(ppp, np)
donorgenerator = DonorGenerator(F)
trialgenerator = TrialGenerator(Cr,problemnumber)
selector = Selector(ppp)
de = DE


DifferentialEvolution = de(ppp, initializer, donorgenerator, trialgenerator, selector)

def visualize(DE,iterations,title='Differential Evolution'):
    results,time=DE.run(iterations)
    fig,ax = plt.subplots(figsize=(8,8))
    for x in range(np):
        plt.plot(results[:,x],label='Chromosome '+str(x))
    ax.ticklabel_format(style='plain')
    ax.axhline(1514312)
    plt.ylim(1400000, 1550000)
    plt.ylabel("Profit/Fitness")
    plt.xlabel("Iteration")
    print('Time needed:',time)
    #plt.legend()
    plt.show()
    
```

Wie ihr sehen könnt sehen die Ergebnisse sehr unterschiedlich aus. Ich lasse das Programm dreimal genau gleich laufen aber die ergebnisse fluktuieren sehr stark. Das könnte an den unterschiedlichen Initialisierungswerten liegen. Allerdinks fluktuieren die erreichbaren Maxima ebenfalls extrem, manchmal erreicht man ein Plateu bei weniger als 2 Mio Profit ein anderes Mal bis über 250 Mio.

Das ist vielleicht ein normales Verhalten des Programmes aber vielleicht auch nicht. U.u. könnte das auf Probleme mit der Profit Funktion, dem Lern-Algorithmus oder etwas anderem schließen. Wie viel Profit sollte man den Maximal überhaupt rechnen können

```python
visualize(DifferentialEvolution,iterations)
visualize(DifferentialEvolution,iterations)
visualize(DifferentialEvolution,iterations)
```
